```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("ufc-fighters-stats.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['stance'] = df['stance'].fillna("Unknown")
df['reach_in_cm'] = df['reach_in_cm'].fillna(df['reach_in_cm'].median())
df['height_cm'] = df['height_cm'].fillna(df['height_cm'].median())
df['weight_in_kg'] = df['weight_in_kg'].fillna(df['weight_in_kg'].median())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['total_fights'] = df['wins'] + df['losses'] + df['draws'] 
df['win_percentage'] = np.where(df['total_fights'] > 0, df['wins'] / df['total_fights'], np.nan)
```

```python
# Extracted year
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
df['birth_year'] = df['date_of_birth'].dt.year
```

```python
df.drop(columns=['nickname', 'date_of_birth'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['win_percentage'].fillna(0, inplace=True)
```

```python
import datetime

# Current year
current_year = datetime.datetime.now().year  

# Calculate Age
df["age"] = current_year - df["birth_year"]

df = df.drop(columns=["birth_year"])
```

```python
df.isna().sum()
```

```python
def assign_weight_class(weight):
    if weight <= 56.7:
        return "Flyweight"
    elif weight <= 61.2:
        return "Bantamweight"
    elif weight <= 65.8:
        return "Featherweight"
    elif weight <= 70.3:
        return "Lightweight"
    elif weight <= 77.1:
        return "Welterweight"
    elif weight <= 83.9:
        return "Middleweight"
    elif weight <= 93.0:
        return "Light Heavyweight"
    elif weight <= 120.2:
        return "Heavyweight"
    else:
        return "Super Heavyweight"
    
df["weight_class"] = df["weight_in_kg"].apply(assign_weight_class)
```

```python
sns.countplot(data=df, y="weight_class", order=df["weight_class"].value_counts().index, palette="viridis")
plt.title("Fighter Distribution by Weight Class (in kg)")
plt.xlabel("Number of Fighters")
plt.ylabel("Weight Class (kg)")
```

```python
cols_to_plot = ["wins","losses","draws","height_cm","weight_in_kg","reach_in_cm","total_fights","win_percentage","age"]

plt.figure(figsize=(8,6))
sns.heatmap(df[cols_to_plot].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Key UFC Fighter Stats)")
```

```python
sns.barplot(data=df, x="weight_class", y="age", estimator=np.mean, palette="coolwarm")
plt.title("Average Age per Weight Class")
plt.xticks(rotation=45)
```

```python
stance_counts = df['stance'].value_counts().sort_values(ascending=False)
sns.barplot(x=stance_counts.index, y=stance_counts.values, palette='viridis')
plt.title('Distribution of Fighter Stances')
plt.xlabel('Stance')
plt.ylabel('Number of Fighters')
plt.xticks(rotation=45)
```

```python
birth_year_counts = df.groupby('age').size().reset_index(name='fighter_count')
plt.figure(figsize=(15, 7))
sns.lineplot(x='age', y='fighter_count', data=birth_year_counts, marker='o')
plt.title('Temporal Trend of Fighters')
plt.xlabel('Age')
plt.ylabel('Number of Fighters')
```

```python
# Showing experience effect on win rate
sns.lineplot(x="age", y="win_percentage", data=df, ci=None, marker="o")
plt.title("Win Percentage vs Age")
plt.xlabel("Age")
plt.ylabel("Win Percentage (%)")
```

```python
# Which stance has highest win rate (very much similar)
sns.barplot(x="stance", y="win_percentage", data=df, ci=None, order=df['stance'].value_counts().index)
plt.title("Average Win Percentage by Stance")
plt.xlabel("Stance")
plt.ylabel("Win Percentage (%)")
```

```python
# getting top players based on win percentage
top_fighters = df.sort_values(by="win_percentage", ascending=False).head(10)
print(top_fighters[['name', 'total_fights', 'win_percentage']])
```

```python
top_strikers = df.sort_values(by="significant_strikes_landed_per_minute", ascending=False).head(10)

# plotting this
sns.barplot(x='significant_strikes_landed_per_minute', y='name', data=top_strikers, palette='viridis')
plt.title('Top 10 UFC Fighters by Significant Strikes Landed per Minute')
plt.xlabel('Significant Strikes Landed per Minute')
plt.ylabel('Fighter Name')
```

```python
df["effective_striking_score"] = (
    df["significant_strikes_landed_per_minute"] * (df["significant_striking_accuracy"] / 100)
)

# Rank fighters based on this combined score
top_effective_strikers = df.sort_values(by="effective_striking_score", ascending=False).head(10)

# plotting this
sns.barplot(x='effective_striking_score', y='name', data=top_effective_strikers, palette='viridis')
plt.title('Top 10 UFC Fighters by Effective Striking Score')
plt.xlabel('Effective Striking Score')
plt.ylabel('Fighter Name')
```

```python
top_defenders = df[df["significant_strikes_absorbed_per_minute"] > 0] \
                    .sort_values(by="significant_strikes_absorbed_per_minute", ascending=False) \
                    .head(10)
# plotting this
sns.barplot(x='significant_strikes_absorbed_per_minute', y='name', data=top_defenders, palette='viridis')
plt.title('Top 10 UFC Fighters by Significant Strikes Absorbed per Minute')
plt.xlabel('Significant Strikes Absorbed per Minute')
plt.ylabel('Fighter Name')
```

```python
df["effective_defense_score"] = (df["significant_strike_defence"] * df["significant_strikes_absorbed_per_minute"])*100

# Top 10 fighters by effective striking
top_effective = df.sort_values(by="effective_defense_score", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x="effective_defense_score", y="name", data=top_effective, palette="crest")
plt.title("Top 10 Fighters by Effective Defense (Defense vs Absorption Balance)")
plt.xlabel("Effective Defense Score")
plt.ylabel("Fighter")
plt.show()
```

```python
df["overall_fighter_score"] = (
    (df["effective_striking_score"] * df["effective_defense_score"])/100
)

# Top 10 overall fighters
top_fighters = df.sort_values(by="overall_fighter_score", ascending=False).head(10)

sns.barplot(x="overall_fighter_score", y="name", data=top_fighters, palette="mako")
plt.title("Top 10 Well-Rounded Fighters (Effective Offense + Defense)")
plt.xlabel("Overall Fighter Score")
plt.ylabel("Fighter")
```

```python
df["grappling_score"] = df["average_takedowns_landed_per_15_minutes"] + df["average_submissions_attempted_per_15_minutes"]

top_grapplers = df.sort_values(by="grappling_score", ascending=False).head(10)

# plotting this
sns.barplot(x='grappling_score', y='name', data=top_grapplers, palette='viridis')
plt.title('Top 10 UFC Fighters by Grappling Score')
plt.xlabel('Grappling Score')
plt.ylabel('Fighter Name')
```

```python
df["defense_score"] = (df["significant_strike_defence"] * df["significant_strikes_absorbed_per_minute"])/100
top_defenders = df.sort_values(by="defense_score", ascending=False).head(10)
# plotting this
sns.barplot(x='defense_score', y='name', data=top_defenders, palette='viridis')
plt.title('Top 10 UFC Fighters by Defense Score')
plt.xlabel('Defense Score')
plt.ylabel('Fighter Name')
```

```python
# Win %
df["total_fights"] = df["wins"] + df["losses"] + df["draws"]
df["win_percentage"] = (df["wins"] / df["total_fights"]) * 100

# Composite score
df["GOAT_score"] = (df["win_percentage"] * 0.4 +
                    df["effective_striking_score"] * 0.2 +
                    df["effective_defense_score"] * 0.2 +
                    df["grappling_score"] * 0.2)

top_goats = df[df["total_fights"] >= 5].sort_values(by="GOAT_score", ascending=False).head(10)

sns.barplot(x='GOAT_score', y='name', data=top_goats, palette='viridis')
plt.title('Top 10 UFC Fighters by GOAT Score')
plt.xlabel('GOAT Score')
plt.ylabel('Fighter Name')
```

```python
from sklearn.preprocessing import MinMaxScaler

# Win %
df["total_fights"] = df["wins"] + df["losses"] + df["draws"]
df["win_percentage"] = (df["wins"] / df["total_fights"]) * 100

# Scale relevant stats
scaler = MinMaxScaler()
stats_to_scale = ["win_percentage", "effective_striking_score", "effective_defense_score", "grappling_score"]

df_scaled = df.copy()
df_scaled[stats_to_scale] = scaler.fit_transform(df[stats_to_scale]) * 100  # scale 0100

# Composite GOAT score with weights (after scaling)
df_scaled["GOAT_score"] = (df_scaled["win_percentage"] * 0.4 +
                           df_scaled["effective_striking_score"] * 0.2 +
                           df_scaled["effective_defense_score"] * 0.2 +
                           df_scaled["grappling_score"] * 0.2)

# Top 10 fighters with min. 5 fights
top_goats_scaled = df_scaled[df_scaled["total_fights"] >= 5] \
    .sort_values(by="GOAT_score", ascending=False) \
    .head(10)

# Plot
sns.barplot(x="GOAT_score", y="name", data=top_goats_scaled, palette="viridis")
plt.title("Top 10 UFC Fighters by Scaled GOAT Score")
plt.xlabel("GOAT Score")
plt.ylabel("Fighter Name")
plt.show()
```
